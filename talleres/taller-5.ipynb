los datos estan dados por:


```
y = cX^n
ln(y) = ln(c) + n * ln(x)

c = e^ln(c)
n = ln(y) - ln(c) / ln(x)
```

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
````

````python
# Lectura del documento
fileLoad = np.loadtxt("taller-5-datos.csv")

# Recuperar todos los elemetos de cada columna
x = fileLoad[:, 0] # columna 1
y = fileLoad[:, 1] # columna 2
````

````python
plt.plot(x, y, "o")
plt.show()
````

````python
# Linealiza los datos de acuerdo al modelo inicialemnte establecido
lny = np.log(y)
lnx = np.log(x)
````

````python
# Se determina mediante minimos cuadrados una aproximacion a las constantes
# del modelo linealizado
res = sts.linregress(lnx, lny)

print(res)

# Genera datos de prueba para verificar graficamente la relacion entre los valores
# resultantes de minimos cuadrados y los valoreles del modelo lineal
nx = np.linspace(np.min(lnx), np.max(lnx), 2)
ny = res.slope * nx + res.intercept

# Grafica los datos
plt.plot(lnx, lny, "o")
plt.plot(nx, ny)
plt.show()
````

````python
# Convierte las constantes que estan linealizadas a la forma del modelo original
c = np.exp(res.intercept)
n = res.slope

print(f"Los valores de la constante c={c} y n={n}")
````

````python
# Genera los datos partiendo del modelo calculado para los datos presentados
xfinal = np.linspace(2, 100, 200)
yfinal = c * xfinal ** n

plt.plot(xfinal, yfinal)
plt.plot(x, y, "o")

# ajustar tamaño de la figura (ancho y alto)
plt.rcParams['figure.figsize'] = [10, 10]

plt.show()
````
