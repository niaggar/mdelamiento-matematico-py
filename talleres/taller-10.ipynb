De acuedo con la física newtoniana, el movimiento de un péndulo simple se puede describir mediante la EDO

\begin{equation}
\dfrac{d^2 \theta}{d t^2} + \dfrac{g}{l}\sin\theta  = 0 .
\end{equation}

Esta ecuación es dificil de solucionar de forma exacta por que es NO lineal. Además, como es una ecuación diferencial de segundo orden, tenemos que definif una nueva variable (velocidad angular) como $v = \dfrac{d\theta}{d t}$ y así obtenemos el sistema: 

\begin{eqnarray}
\dfrac{d \theta}{d t} &=& v  \\
\dfrac{d v}{d t} &=& -\dfrac{g}{l}\sin\theta
\end{eqnarray}


El campo de velocidades para este sistema en el plano $(\theta, v)$ en cierto punto $(\theta_0, v_0)$ se puede definir como como 

\begin{equation}
u \big|_{\theta_0,v_0}  = (u_1 , u_2) :=\dfrac{1}{\sqrt{ V_1^2 +  V_2^2 } } ( V_1 ,  V_2).
\end{equation}

donde $ V_1 = \dfrac{d \theta}{dt}\big|_{\theta_0, v_0} $ y  $V_2 = \dfrac{d v}{dt} \big|_{\theta_0, v_0} $. 

Con la informacion anterior, y asumiendo la longitud del péndulo $l= 1$:

- Grafique el campo de velocidades de este sistema para $\theta \in [- \pi, \pi]$ y  $v \in [-2 \pi,2 \pi]$.

- Del campo de velocidades estime el valor máximo que alcanza $\theta$ si se comienza con un dato inicial de $\theta_0 =0$ y $v_0=1$

- Del campo de velocidades estime el valor máximo que alcanza la velocidad $v$ si se comienza con un dato inicial de $\theta_0 =0$ y $v_0=0$

- Resuelva el sistema utilizando el metodo de Euler y grafique la funcion de $\theta$ contra $t$ partiendo de la condicion inicial $(\theta_0, v_0)=(0,1)$. Use un dominio para $t$ en $[0, \pi]$ y un tamaño de paso de $0.0001$. 




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#--------Ecuacion diferencial-----
def Ec_differencial(theta):
    return -(9.81/1) * np.sin(theta)

#---Definiendo parametros---
theta0 = -np.pi
thetaf = +np.pi


vmin = -2 * np.pi
vmax = +2 * np.pi

# Valores para facilitar la estimacion del campo de velocidades
# theta0 = -1
# thetaf = 1
# vmin = -2
# vmax = +2
#---------------------------


#Numero de puntos
N   =  20


#valores de theta
theta  = np.linspace(theta0, thetaf, N)

#valores de la funcion
velocidad  = np.linspace(vmin, vmax, N)

#Creando la malla de puntos (t,Y)
theta_grid, velocidad_grid = np.meshgrid(theta, velocidad)


#-------------Vector de velocidades ------------------
#componente horizontal
Vectorx = velocidad_grid

#componente vertical
Vectory = Ec_differencial(theta_grid)

#Normalizando el vector
Norma = np.sqrt(Vectorx**2 + Vectory**2)

ux = Vectorx / Norma
uy = Vectory / Norma
#----------------------------------------------------

#graficando el campo de velocidades
plt.quiver(theta_grid, velocidad_grid, ux, uy )
plt.title("Campo de velocidades V")
plt.xlabel('Theta')
plt.ylabel('Velocidad')
plt.axis([theta0, thetaf, vmin, vmax])
plt.show()

plt.streamplot(theta_grid, velocidad_grid, ux, uy, density = 1)
plt.title("Campo de velocidades V")
plt.xlabel('Theta')
plt.ylabel('Velocidad')
plt.axis([theta0, thetaf, vmin, vmax])
plt.show()
```

Acorde al campo de velocidades se puede estimar que:


1.   El valor maximo de theta cuando theta0 = 0 y velocidad0 = 1 es de 0.375.
2.   El valor maximo de velocidad cuando theta0 = 0 y velocidad0 = 0 es de 0.




```python
#--------Ecuacion diferencial-----
def Ec_differencial( theta ):
    dvdt = -(9.81/1) * np.sin(theta)
    return dvdt
#---------------------------------

# Resuelva el sistema utilizando el metodo de Euler y grafique la funcion de θ contra t partiendo de la condicion inicial (θ0,v0)=(0,1). 
# Use un dominio para t en [0,π] y un tamaño de paso de 0.0001.


#PARAMETROS
theta0 = 0
thetaf = np.pi
dt = 0.0001

#lista de tiempos  
t = np.arange(theta0, thetaf, dt)

#Numero de tiempos


#lista los valores de la función
theta = np.arange(theta0, thetaf, dt)
velocidad = np.zeros_like(theta)

N = len(theta)

#asigno el valor inicial
velocidad[0] = 1

#Aplicamos la formula de Euler
for i in range( N-1 ):
    theta[i+1] = theta[i] +  dt * theta[i]
    velocidad[i+1] = velocidad[i] +  dt * Ec_differencial(velocidad[i])

plt.plot(theta, velocidad,'-r')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.show()
```
