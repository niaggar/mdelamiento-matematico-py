````python
# Carga de librerias requeridas
import numpy as np
import matplotlib.pyplot as plt
````

1. Leer y graficar los datos

````python
# Lectura del documento
fileLoad = np.loadtxt("datos.dat")

# fileLoad[0,0]     --- Elemento de fila 0 columna 0
# fileLoad[0:4, 0]  --- Elementos desde fila 0 hasata fila 4 y columna 0

# Recuperar todos los elemetos de cada columna
x = fileLoad[:, 0] # columna 1
y = fileLoad[:, 1] # columna 2

print(fileLoad)
````

````python
# Asignar grafico
plt.plot(x, y, marker='D', color="g")

# Establecer labels
plt.ylabel("Valores recuperados de y")
plt.xlabel("Valores recuperados de x")

# ajustar tama√±o de la figura (ancho y alto)
plt.rcParams['figure.figsize'] = [1, 1]

# Mostrar grafico
plt.show()
````

2. Realacion de proporcionalidad

Es una proporcionalidad inversa


```
y = k / x
y * x = K

y = k * x
y / x = k
```


Ya que conforme los valores que toma *x* aumentan los valores que toma *y* disminuyen

3. Determine constante de proporcionalidad

````python
# Declaracion constante k
k = 0

# Multiplicacion de todos los valores de x por todos los valores de y para
# determinar k en cada punto de la grafica
array_k = y / x
# [a, b, c] * [a2, b2, c2] = [a*a2, b*b2, c*c2]

# Calculo del promedio de la constante k con la utilizacion de la funcion
# average integrada en numpy, la cual ya esta definida y automaticamente
# suma los valores del arreglo y los divide por la longitud del propio arreglo
k = np.average(array_k)

sum(array_k)/len(array_k)

print(k)
````
