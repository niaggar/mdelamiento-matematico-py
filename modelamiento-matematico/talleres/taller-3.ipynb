```python
import matplotlib.pyplot as plt
```

```python
a = 2
b = 2
c = 2
d = 2
```

```python
def polinomio(x, a, b, c, d):
  p = (a * (x**3)) + (b * (x**2)) + (c * x) + d

  return p
```

```python
xMin = -20
xMax = 20

xValues = list(range(xMin, xMax + 2, 2))
yValues = []

for x in xValues:
  y = polinomio(x, a, b, c, d)
  yValues.append(y)
```

```python
plt.plot(xValues, yValues, "ro")
plt.plot(xValues, yValues, "b--")

plt.ylabel("Valores de ax^3 + bx^2 + cx + d")
plt.xlabel("Valores de x par")

plt.show()
```
